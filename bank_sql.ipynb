# Analyse SQL - Bank Marketing

```python
import pandas as pd
import sqlite3

print("outils importes")
```

```python
df = pd.read_csv('bank_debutant.csv')
conn = sqlite3.connect('bank_database.db')
df.to_sql('clients', conn, if_exists='replace', index=False)
print("base de donnees creee")
```

```python
# fonction pour ecrire du sql 
def sql(requete):
    return pd.read_sql_query(requete, conn)
```

```python
sql("SELECT y, COUNT(*) AS nombre FROM clients GROUP BY y")
```

```python
# age moyen et solde moyen
sql("SELECT ROUND(AVG(age), 1) AS age_moyen, ROUND(AVG(balance), 0) AS solde_moyen FROM clients")
```

```python
# nombre de clients par metier
sql("""
SELECT job AS metier, COUNT(*) AS nombre_clients
FROM clients
GROUP BY job
ORDER BY nombre_clients DESC
""")
```

```python
# solde moyen selon souscription
sql("""
SELECT y, ROUND(AVG(balance), 0) AS solde_moyen
FROM clients
GROUP BY y
""")
```

```python
# taux de souscription par metier
sql("""
SELECT job AS metier,
COUNT(*) AS total,
SUM(CASE WHEN y = 'yes' THEN 1 ELSE 0 END) AS nb_souscrit,
ROUND(SUM(CASE WHEN y = 'yes' THEN 1 ELSE 0 END) * 100.0 / COUNT(*), 1) AS taux_pct
FROM clients
GROUP BY job
ORDER BY taux_pct DESC
""")
```

```python
# taux de souscription par mois
sql("""
SELECT month AS mois,
COUNT(*) AS total,
SUM(CASE WHEN y = 'yes' THEN 1 ELSE 0 END) AS nb_souscrit,
ROUND(SUM(CASE WHEN y = 'yes' THEN 1 ELSE 0 END) * 100.0 / COUNT(*), 1) AS taux_pct
FROM clients
GROUP BY month
ORDER BY taux_pct DESC
""")
```

```python
# canal de contact le plus efficace
sql("""
SELECT contact AS canal,
COUNT(*) AS total,
ROUND(SUM(CASE WHEN y = 'yes' THEN 1 ELSE 0 END) * 100.0 / COUNT(*), 1) AS taux_pct
FROM clients
GROUP BY contact
ORDER BY taux_pct DESC
""")
```

```python
# profil du client qui a souscrit
sql("""
SELECT y,
ROUND(AVG(age), 1) AS age_moyen,
ROUND(AVG(balance), 0) AS solde_moyen,
ROUND(AVG(duration), 0) AS duree_appel_moy
FROM clients
GROUP BY y
""")
```
